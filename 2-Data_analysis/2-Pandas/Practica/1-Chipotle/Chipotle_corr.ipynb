### Step 4.2: Y cuÃ¡ntos pedidos se han hecho de mÃ¡s de 10$? Es lo mismo?

```python
pedidos = chipo.groupby("order_id")["item_price"].sum()
pedidos_10 = pedidos[pedidos >10]
print(len(pedidos_10))
```

### Step 4.3: Y en cuÃ¡ntos pedidos se ha pagado mÃ¡s de 10$ por un mismo producto? Es lo mismo?

```python
var = chipo[chipo["item_price"] > 10].groupby(["order_id", "item_name"]).size()
print(len(var))
```

```python
print(len(chipo[chipo["item_price"] > 10].groupby(["order_id", "item_name"]).size()))
```

```python
pedidos = chipo.groupby("order_id")["item_price"].sum()
pedidos_10 = pedidos[pedidos >10]
print(len(pedidos_10))
```
