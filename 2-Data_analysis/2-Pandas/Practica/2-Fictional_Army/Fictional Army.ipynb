# Fictional Army - Filtering and Sorting

```python
# Create an example dataframe about a fictional army
raw_data = {'regiment': ['Nighthawks', 'Nighthawks', 'Nighthawks', 'Nighthawks', 'Dragoons', 'Dragoons', 'Dragoons', 'Dragoons', 'Scouts', 'Scouts', 'Scouts', 'Scouts'],
            'company': ['1st', '1st', '2nd', '2nd', '1st', '1st', '2nd', '2nd','1st', '1st', '2nd', '2nd'],
            'deaths': [523, 52, 25, 616, 43, 234, 523, 62, None, 73, 37, 35],
            'battles': [5, 42, 2, 2, 4, 7, 8, 3, 4, 7, 8, 9],
            'size': [1045, 957, 1099, 1400, 1592, 1006, 987, 849, 973, 1005, 1099, 1523],
            'veterans': [1, 5, 62, 26, 73, 37, 949, 48, 48, 435, 63, 345],
            'readiness': [1, None, 3, None, 2, None, 2, None, None, 1, 2, None],
            'armored': [1, 0, 1, 1, 0, 1, 0, 1, 0, 0, 1, 1],
            'deserters': [4, 24, 31, 2, 3, 4, 24, 31, 2, 3, 2, 3],
            'origin': ['Arizona', 'California', 'Texas', 'Florida', 'Maine', 'Iowa', 'Alaska', 'Washington', 'Oregon', 'Wyoming', 'Louisana', 'Georgia']}
```

### Import libraries

### Step 1. This is the data given as a dictionary. Create a dataframe and assign it to a variable called army. 

```python
import numpy as np
import pandas as pd

df_army = pd.DataFrame(raw_data)
df_army.head()
```

```python
df_army.info()
```

```python
df_army.shape
```

```python
df_army.describe(include='all')
```

### Step 2. Check missing values

```python
missings = df_army.isnull().sum()
print(missings)
```

### Step 3. Drop features, if there are more than 30% missing values.

```python
threshold = 0.3 * len(df_army)
threshold
```

```python
cols_2_drop = missings [missings > threshold].index.values
# drop features with more than 30% missings
df_army.drop(columns=cols_2_drop, inplace=True)
df_army.head()
```

```python
df_army.shape
```

```python
df_army.info()
```

* alternativa `dropna()`

```python
len(df_army) * 0.7
```

```python
df_army =df_army.dropna(axis = 1, thresh= len(df_army) * 0.7)
df_army
```

### Step 4. Fill missing values with the mean of their regiment.

```python
df_army[df_army.regiment == 'Scouts']
```

```python
scouts_deaths_mean = round(df_army[df_army.regiment == 'Scouts'].groupby('regiment')['deaths'].mean().iloc[0], 1)
scouts_deaths_mean
```

```python
index_nan = df_army[df_army["deaths"].isna()].index.values[0]       # aqui es el 8
df_army.loc[index_nan,'deaths'] = scouts_deaths_mean
```

```python
df_army.loc[8]
```

```python
# df_army.deaths = df_army.deaths.fillna(valor)

df_army = df_army.deaths.fillna(df_army.groupby('regiment')['deaths'].transform('mean'))            # forma AUTO
```
