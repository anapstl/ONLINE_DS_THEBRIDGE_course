2. Sabemos que podemos pintar gráficas de dos formas: la figura incluye los ejes o teniendo figura y ejes por separado.

Usando solo una figura, usa numpy para los valores del eje X entre 0 y 5. 

Pinta dos gráficas en dos cajas distintas, a la izquierda una recta con pendiente positiva de 3 que pase por (0,0) y a la derecha una recta con pendiente negativa de 3 que pase por (0,-5). 

Elige la precisión en el eje X que desees.  

```python
# Usando solo una figura, usa numpy para los valores del eje X entre 0 y 5. 
fig = plt.figure()
ax = plt.axes()

x = np.linspace(0, 5, 100)
ax.plot(x);
```

```python
# Pinta dos gráficas en dos cajas distintas, a la izquierda una recta con pendiente 
# positiva de 3 que pase por (0,0) y a la derecha una recta con pendiente negativa de 3
#  que pase por (0,-5). 

# Elige la precisión en el eje X que desees.  

fig, axs = plt.subplots(nrows=1, ncols=2, layout='constrained', figsize=(10, 4))
x = np.linspace(-10, 10, 100)

y1 = 3 * x                 # pendiente +3, pasa por (0, 0)
y2 = -3 * x - 5            # pendiente -3, pasa por (0, -5)

for nn, ax in enumerate(axs):
    ax.set_xlabel("eje X")
    ax.set_ylabel("eje Y")
    ax.axhline(0, color="red", linewidth=0.25)
    ax.axvline(0, color="red", linewidth=0.25)
    if nn == 1:
        ax.set_title('Automatic ticks/ Recta baja')
        plt.subplot(1, 2, 1);
        plt.plot(x, y2, label='y = 3x', color='blue');
    else:
        ax.set_title('Manual ticks/ Recta sube')
        plt.subplot(1, 2, 2);
        ax.set_xticks(np.arange(-10, 10, 10/10))
        plt.plot(x, y1, label='y = -3x - 5', color='red');
```

5. Usando una figura Y EJES POR SEPARADO, usa numpy para los valores del eje X entre 0 y 5. Pinta dos gráficas en dos cajas distintas, a la izquierda una recta con pendiente positiva de 3 que pase por (0,0) y a la derecha una recta con pendiente negativa de 3 que pase por (0,-5). Elige la precisión en el eje X que desees.  

```python
x = np.linspace(0, 5, 100)

fig = plt.figure(figsize=(10, 4))

ax1 = fig.add_axes([0.1, 0.2, 0.35, 0.6])
ax1.plot(x, 3 * x, 'b')  # y = 3x
ax1.set_xticks(np.arange(0, 5, 5/10))
ax1.set_title('Recta Positiva')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')

ax2 = fig.add_axes([0.55, 0.2, 0.35, 0.6])
ax2.plot(x, -3 * x - 5, 'r')  # y = -3x - 5
ax2.set_title('Recta Negativa')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')

plt.show()
```

13. Vamos a introducir pandas. Crea un DataFrame con las columnas "Year" de valores 2015, 2016, 2017, 2018, 2019 y la columna "Sold_items_A" de valores 1000, 3500, 4000, 5500, 7000

```python
year = [2015, 2016, 2017, 2018, 2019]
sold_items_A = [1000, 3500, 4000, 5500, 7000]

df = pd.DataFrame({
    'Year': year,
    'Sold_items_A':sold_items_A
})

df
```

14. Ahora pintamos un gráfico de línea con las ventas respecto al año. Pon un título y etiquetas en los ejes.
Cambia el estilo a 'seaborn-white'. La línea debe ser con rayas y verde.

```python
plt.style.use('seaborn-v0_8-white')

fig = plt.figure()
ax = plt.axes()

ax.plot(df.Year.astype(int), df.Sold_items_A, '--g');
# ax.set_xticklabels(df['Year'].astype(int))
plt.title('Ejercicio 14')
plt.ylabel('Sold Items A')
plt.xlabel('Year')
```

15. Otro departamento B ha vendido en esos años 2000, 3100, 5000, 4000, 6000 unidades. Incluye esa columna en el DataFrame y pinta en la misma gráfica las dos líneas. B es una línea punteada y roja. Muestra una leyenda abajo a la derecha.

```python
b = [2000, 3100, 5000, 4000, 6000]
df['Sold_items_B'] = b
df
```

```python
plt.style.use('seaborn-v0_8-white')

fig = plt.figure()
ax = plt.axes()

ax.plot(df.Year.astype(int), df.Sold_items_A, '--g', label = 'Sold Items A');
ax.plot(df.Year.astype(int), df.Sold_items_B, ':r', label = 'Sold Items B');
ax.legend(loc='lower right')
# ax.set_xticklabels(df['Year'].astype(int))
plt.title('Ejercicio 15')
plt.ylabel('Sold Items')
plt.xlabel('Year')
plt.show()
```
