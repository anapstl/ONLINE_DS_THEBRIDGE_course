# Constructores

URL: https://colab.research.google.com/drive/1Ny1wJh4djAH0XZJurj4qYTmE5HdCSjpB

crear obj. con sus attrib. con valores diferenciados - `constructores de clase` mtodo similar a una fct al que daremos unos args de entrada que sustituirn a los vals x defecto y nos permitirn diferenciar esa instancia de otras instancias de la misma clase

funcion `__ini__`

```python
class Coche:
    puertas = 4
    ruedas = 4
    def __init__(self, marca_coche, modelo):
        self.marca_coche = marca_coche
        self.modelo = modelo
```

```python
tesla = Coche(marca_coche="Tesla", modelo="Model S")
print(tesla.marca_coche)
```

```python
mercedes = Coche(marca_coche="Mercedes", modelo="Clase A")
renault = Coche(marca_coche="Renault", modelo="Clio")
peugeot = Coche(marca_coche="Peugeot", modelo="308")
print(mercedes.marca_coche)
print(renault.marca_coche)
print(peugeot.marca_coche)
print(mercedes.puertas)
print(renault.puertas)
print(peugeot.puertas)
```

```python
# modificar

class Coche():
    puertas = 4
    ruedas = 4
    def __init__(self, marca_coche, modelo, *num_puertas):
        self.marca_coche = marca_coche
        self.modelo = modelo
        self.num_puertas = num_puertas if num_puertas else self.puertas
```

```python
tesla = Coche("Tesla", "Model S", 8)
mercedes = Coche("Mercedes", "Clase A", 22)
print(tesla.num_puertas)
print(mercedes.num_puertas)
```

```python
peugeot = Coche("Peugeot", "308")
print(peugeot.num_puertas)
```

```python
# modificar 2

class Coche():
    # puertas = 4
    # ruedas = 4
    def __init__(self, marca_coche, modelo, num_puertas = 4, num_ruedas = 4):
        self.marca_coche = marca_coche
        self.modelo = modelo
        self.num_puertas = num_puertas
        self.num_ruedas = num_ruedas
```

```python
tesla_4 = Coche("Tesla", "Model S")
tesla_6r = Coche("Tesla", "Model 6r", num_ruedas=6)

print(tesla_4.num_ruedas)
print(tesla_6r.num_ruedas)

print(tesla_4.num_puertas)
print(tesla_6r.num_puertas)
```
