```python
import pandas as pd
import numpy as np
from sklearn import linear_model
from sklearn import model_selection
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

## Creamos el modelo

```python
X = np.array(dataframe.drop(columns=['clase']))
y = np.array(dataframe['clase'])
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(max_iter=1000)
model.fit(X, y)
```

```python
predictions = model.predict(X)
print(predictions)
```

```python
predicions_proba = model.predict_proba(X)
print(np.round(np.array(predicions_proba), 2))
```

```python
model.score(X, y)
```

```python
model.classes_
```

# Validación del Modelo

```python
validation_size = 0.20
seed = 7
X_train, X_test, Y_train, Y_test = model_selection.train_test_split(X,
                                                                    y,
                                                                    test_size=validation_size,
                                                                    random_state=seed)
```

```python
name='Logistic Regression'
kfold = model_selection.KFold(n_splits=10) #Parte los datos en 10 trozos para usar validación cruzada / cross validation
cv_results = model_selection.cross_val_score(model, X_train, Y_train, cv=kfold, scoring='accuracy')

msg = "%s: %f (%f)" % (name, cv_results.mean(), cv_results.std())
print(cv_results)
print(msg)
```

```python
from sklearn.model_selection import GridSearchCV

params = {"C" : np.arange(0.1,10,0.1)}

gs = GridSearchCV(estimator=model, param_grid=params, cv=3, scoring='accuracy', verbose=3)
gs.fit(X_train, Y_train)

print(gs.best_estimator_)
print(gs.best_score_)
print(gs.best_params_)
```

```python
model = gs.best_estimator_ # res del gridsearch y es equiv de la siguiente linea, asignacion manual del modelo
# model = LogisticRegression(max_iter=1000) 
model.fit(X_train, Y_train)
```

```python
predictions = model.predict(X_test)
print(accuracy_score(Y_test, predictions))
```

```python
acierto = accuracy_score(Y_test, predictions)

error = 1 - acierto
print("Acierto:", round(acierto*100, 2), "%")
print("Error:", round(error*100, 2), "%")
```
