Def. problema **"Gestor biblio"**

- añadir libro
- eliminar libro
- buscar un libro x titulo
- mostrar todos los libros

## Resolución funcional

```python
def anadir_libro(biblio, libro):
    biblio.append(libro)

def eliminar_libro(biblio, titulo):
    biblio[:] = [libro for libro in biblio if libro['titulo'] != titulo]

def buscar_libro(biblio, titulo):
    for libro in biblio:
        if libro['titulo'] == titulo:
            return libro
    return None

def mostrar_libros(biblio):
    print("Monstrando libros biblio:")
    for libro in biblio:
        print("- ", libro['titulo'], libro['autor'])
    return None
```

```python
# Ejemplo de uso
biblio = []
anadir_libro(biblio, {'titulo':'1984', 'autor':'george orwell'})
anadir_libro(biblio, {'titulo':'brave new world', 'autor':'aldous huxley'})
print(biblio)
print("Busco 1984:", buscar_libro(biblio, "1984"))
print("-----")
mostrar_libros(biblio)
print("-----")
print("Elimino 1984:", eliminar_libro(biblio, "1984"))
print("-----")
mostrar_libros(biblio)
print("-----")
```

## Resolución POO

```python
class Libro:
    
    def __init__(self, titulo_libro, autor_libro):
        self.titulo = titulo_libro
        self.autor = autor_libro

    def print_datos(self):
        print('titulo:', self.titulo)
        print('autor:', self.autor)
        return self.titulo
```

```python
class Biblio:
    def __init__(self):
        self.libros = []
    
    def añadir_libro(self, libro):
        self.libros.append(libro)

    def eliminar_libro(self, titulo):
        self.libros = [libro for libro in self.libros if libro.titulo != titulo]

    def buscar_libro(self, titulo):
        for libro in self.libros:
            if libro.titulo == titulo:
                libro.print_datos()
                return libro
        return None
    
    def mostrar_libros_poo(self):
        for libro in self.libros:
            libro.print_datos()
```

```python
mi_biblio = Biblio()
l_1984 = Libro("1984", "G.Orwell")
mi_biblio.añadir_libro(l_1984)
bnw = Libro("Brave New World", "A. Huxley")
mi_biblio.añadir_libro(bnw)
mi_biblio.mostrar_libros_poo()
```

```python
mi_biblio.buscar_libro("1984")
```

```python
mi_biblio.eliminar_libro("1984")
mi_biblio.mostrar_libros_poo()
```
